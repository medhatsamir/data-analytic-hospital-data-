## Introduction

This dataset collects information from 100k medical appointments in Brazil and is focused on the question of whether or not patients show up for their appointment. A number of characteristics about the patient are included in each row.
1-‘ScheduledDay’ tells us on what day the patient set up their appointment.
2-‘Neighborhood’ indicates the location of the hospital.
3-‘Scholarship’ indicates whether or not the patient is enrolled in Brasilian welfare program Bolsa Família.
4-In column ‘No_show’ it says ‘No’ if the patient showed up to their appointment, and ‘Yes’ if they did not show up.

## Data Wrangling

import the libraries 

```python
# import the libraries  
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline 
```

load data and print the head of it 

```python
df = pd.read_csv("noshowappointments-kagglev2-may-2016.csv")
df.head(5)
```

describe for this data 

```python
df.describe()
```

```python
#check null values 
df.info()
print('*'*50)
#check the duplication 
print('there is : ' + str(df.duplicated().sum()) + ' duplicated row' + '\n'+ 'there is : '+ str(df.isnull().all().sum()) +' Null values')

```


### Data Cleaning
 
 columns changing  

```python
# make 'ScheduledDay' and 'AppointmentDay' datetime
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay']  = pd.to_datetime(df['AppointmentDay'])

df.head()
```

```python
# check the 0 age or less
df[df['Age']<= 0]
```

```python
#fixing the 0 age or less
df[ df['Age'] <= 0 ] = df.Age.mean()
df[ df['Age'] <= 0 ]
```

```python
# make the name of the coulmns much easier by using _
new_labels = ['patient_ID', 'appointment_ID', 'gender', 'schedule_day', 'appointment_day', 'age', 'neighborhood',
              'scholarship', 'hypertension', 'diabetes', 'alcoholism', 'handicap', 'sms_received', 'no_show']
df.columns = new_labels
df.head(5)
```

```python
# convert yes to 1 and no to 0 in no_show data 
df.no_show[df['no_show'] == 'Yes'] = 1
df.no_show[df['no_show'] == 'No'] = 0
showed = df['no_show'] == 0
not_showed = df['no_show'] == 1
df.head()
```

```python
# append new coulmn to each showed and not showed 
showed = df['no_show'] == 0
not_showed = df['no_show'] == 1

df['showed'] = showed
df['not_showed'] = not_showed
df.head()
```

```python
df[df['showed'] == df['not_showed']].head()
```

```python
# drop the row which showed and not showed at the same time 
df.drop(df[df['showed'] == df['not_showed'] ].index, inplace = True)
```

```python
df[df['showed'] == df['not_showed']].head()
```

## Exploratory Data Analysis




### Question 1 (Proportion Of Appointment Showed VS Not Showed?

```python
# how many patient 
print('there are :  {}  patients'.format(df.shape[0]) )
```

```python
# plotting 
pieChart = df.not_showed.value_counts().plot.pie(figsize = (8 ,8 ) , autopct='%11.1f%%' ,labels = ['Showed', 'Not Showed'] , explode = [0.1,0])
pieChart.set_title('Showed and Not Showed status', fontsize = 12)
plt.legend()
```

### Question 2 : what is the most important attribute ('Age', 'being alcoholic','Having an SMS', 'Gender', 'Scholarship'), and make the patient show ?

```python
# average age of people who shwed and not showed 
age_Showed = df.age[showed].mean()
age_Not_Showed = df.age[not_showed].mean()
print('the averge age of people who showed is :' + str(age_Showed) + "\n" + 'the averge age of people who not showed is : ' + str(age_Not_Showed))
```

Plot the relation between Age and patients who showed

```python
df.groupby('age')['showed'].mean().hist(alpha = 0.7 , bins=25 ,label='show', color='blue')
df.groupby('age')['not_showed'].mean().hist(alpha=0.7 ,bins=25, label='not show',color='orange');
plt.ylabel("Age")
plt.title("mean Age of people who showed Vs not showed")
plt.legend();
```

plot the relation between Alcoholism and patients who showed


```python
df.groupby('alcoholism')['showed'].mean().plot(kind='pie',figsize = (6 ,5 ) , explode =[0.1,0], autopct='%11.1f%%' , labels = ['showed', 'not showed' ])
plt.ylabel('Alcoholism')
plt.title("relation between alcoholics and patients who showed")
plt.legend();
```

Plot the relation between sms_received and patients who showed¶


```python
df.groupby('sms_received')['showed'].mean().plot(kind='pie',figsize = (6 ,5 ) ,explode =[0.1,0] ,labels = ['showed', 'not showed' ], autopct='%11.1f%%')

plt.ylabel('sms_received')
plt.title("relation between sms_received and patients who showed")
plt.legend();
```

Plot the relation between Gender and patients who showed

```python
df.groupby('gender')['showed'].mean().plot(kind='pie' ,figsize = (6 ,5 ) ,explode =[0.1,0]  ,labels = [ 'FeMale' ,'male'] , autopct='%11.1f%%')
plt.xlabel("Showed")
plt.ylabel("gender")
plt.title("relation between Gender and patients who showed")
plt.legend();
```

Plot the relation between Scholarship and patients who showed

```python
df.groupby('scholarship')['showed'].mean().plot(kind='pie' ,figsize = (6 ,5 ) ,explode =[0.1,0] ,labels = ['showed', 'not showed' ] , autopct='%11.1f%%')
plt.ylabel("scholarship")
plt.title("relation between Scholarship and patients who showed")
plt.legend();
```

figure the relation between neighborhood and patients who showed

```python
df.groupby('neighborhood')['showed'].mean().plot(kind='barh' ,figsize=(20,20) , colormap='Paired' )
plt.xlabel("Showed")
plt.ylabel("scholarship")
plt.title("relation between neighborhood and patients who showed")
plt.legend();
```

## Conclusions

• The most crucial factor that determines whether a patient will show or not is their age; the average age of those who are most likely to show is 39, while the average age of those who are not likely to show is 35.

• Gender, alcoholism, or sending a text message for an appointment has no bearing on getting the patient to show up.

• 22.8 percent of people who scheduled an appointment but didn't show up, implying that the 1/4 of people are most likely to show up late.

### Limitations
More data is missing that might be more beneficial in predicting the patient's show, such as weather, employment, or regional data.

There were some nonsensical facts, such as patients who were under the age of one.
